```python
#importing libraries
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori
from sklearn.preprocessing import Imputer
```

```python
#read and load csv file
data_module= pd.read_csv('Market_Basket_Optimisation.csv',header=None)
data_module.head()
data_module.shape
```

```python
#replace missing value by mean
imp = Imputer(missing_values=0, strategy='mean', axis=0)
```

```python
transaction=[]
for i  in range(0,7501):

#To convert pandas dataframe into list of list.
    transaction.append([str(data_module.values[i,j])for j in range(0,20)])
    
#applying apriori algorithm
    rules = apriori(transaction,min_support=0.003,min_lift=3,min_length=2)
    
```

```python
association_rules = list(rules)
```

```python
print(len(association_rules))
# we have in all 176 rules mined form 7501 transactions
```

```python
#lets see the first rule
association_rules[0]
```

```python
for item in association_rules:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

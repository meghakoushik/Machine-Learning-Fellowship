```python
data_module.replace(np.NaN,data_module.mean(),inplace= True)
data_module.head()
```

```python
data_module.replace(['yes','no'],[1,0],inplace=True)
data_module.head()
#data_module.tail()
```

```python
data_module= pd.get_dummies(data_module)
data_module.head()
```

```python
#feature scalling
data_module = (data_module-data_module.min())/data_module.max()-data_module.min()
data_module.head()
```

```python
#seprating the output
def separate(data_module):
    output = data_module.y
    data_module = data_module.drop('y', axis=1)
    return data_module, output 
data_module, output = separate(data_module)
```

```python
def split(data_module,output):
    train_data = int(0.70*len(data_module))
    test_data = len(data_module)-train_data
    print("train data",train_data)
    print("test_data",test_data)
    x_train = np.array(data_module[:train_data])   
    x_test = np.array(data_module[:test_data])

   
    train_per_y = int(0.70*len(output))
    test_per_y = len(output)-train_per_y
    
    y_train = np.array(output[:train_per_y])
    y_test = np.array(output[:test_per_y])
    
    y_train = y_train.reshape(-1,1)
    y_test = y_test.reshape(-1,1)
    
    return x_train,y_train,x_test,y_test
x_train,y_train,x_test,y_test = split(data_module,output)
print("train data shape",x_train.shape)
print("train data shape",y_train.shape)

print("train data shape",x_test.shape)

print("train data shape",y_test.shape)
```

```python
class KNN_problem:
    def __init__(self):
        self.k=5
    
    
    def Euclidean(self,x_test,x_train,length):
        distance=0
        for i in range(length):
            distance+=pow(x_test[i]-x_train[i],2)
        return math.sqrt(distance)
    
    def neighbours(self,x_train,x_test,y_train):
        distance=[]
        length=len(x_test)-1
        for i in range(len(x_train)):
            dist=self.Euclidean(x_test,x_train[i],length)
            distance.append((y_train[i],dist))
            
        distance.sort(key=operator.itemgetter(1))
        neighbour=[]
        for i in range(self.k):
            neighbour.append(distance[i][0])
        return neighbour
        
        

    def Majority(self,neighbors):
        majority = []
        for x in range(len(neighbors)):
            response = neighbors[x]
            if response in majority:
                majority[response] += 1
            else:
                majority[response] = 1
        majority = sorted(majority.items(), key=operator.itemgetter(1), reverse=True)
        return majority[0][0]
  
    def Accuracy(self,y_test, predictions):
        correct = 0
        for x in range(len(y_test)):
            if y_test[x] == predictions[x]:
                correct += 1
        return (correct/float(len(y_test))) * 100.0

        
def main(x_train,y_train,x_test,y_test):
    obj = KNN_problem()
    # calling method by class object
    
    x_train = np.array(x_train[:2000])
    y_train = np.array(y_train[:2000])
       
    x_test = np.array(x_test[:800])
    y_test = np.array(y_test[:800])
    predictions=[]
    for i in range(len(x_test)):
        neighbours = obj.neighbours(x_train,x_test[i],y_train)
        result = obj.Majority(neighbours)
        
        predictions.append(result)
        print('> predicted=' + repr(result) + ', actual=' + repr(y_test[i]))
        
    accuracy =obj.Accuracy(y_test, predictions)
    print('\nAccuracy: ' + repr(accuracy) + '%')

if __name__ == '__main__':
    main(x_train,y_train,x_test,y_test)
```

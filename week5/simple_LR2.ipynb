```python
#split given dataset into training data and test data
def split(data_module):
    
    x = (data_module["Apparent_Temperature"]) 
    y = (data_module['Humidity'])
    
    train_data = int(0.70*len(data_module))     
    test_data = len(data_module) - train_data 
    
    print("train data", train_data)
    print("Test data", test_data)
    
    train_data_array_x = np.array(x[:train_data]) 
    train_data_array_y =np.array(y[:train_data])
    test_data_array_x = np.array(x[train_data:])
    test_data_array_y = np.array(y[train_data:]) 
    
    return train_data_array_x,train_data_array_y,test_data_array_x,test_data_array_y

train_data_array_x,train_data_array_y,test_data_array_x,test_data_array_y = split(data_module)
train_data_array_x.shape, test_data_array_x.shape
```

```python
class LinearRegression:  
    def __init__(self):
        self.learning_rate = 0.01
        self.epoch = 30

  #find gradient descent
    def gradient_descent(self,theta_0,theta_1,train_data_array_x,train_data_array_y):
        print(train_data_array_x,train_data_array_y)
        length_traindata = len(train_data_array_x)
        for value_item in range(self.epoch):
            h2 = 0
            h1 = 0
            #cost=0
            for i in range (length_traindata):
                hypothesis_function = theta_0 + (theta_1 * train_data_array_x[i])
                h1 += (hypothesis_function - train_data_array_y[i])
                h2 += (hypothesis_function - train_data_array_y[i] * train_data_array_x[i])
                cost = ((hypothesis_function-train_data_array_y[i])**2)
                #values theta_0 and theta_1
            theta_0 = theta_0 - (((self.learning_rate)/len(train_data_array_x))*h1)
            theta_1 = theta_1 - (((self.learning_rate ) / len(train_data_array_x)) * h2)    
            cost_function = (1/2*length_traindata)*(cost)
        print("cost_function", cost_function)
    
        return theta_0, theta_1
        
        #cost_function = (1/2*len(train_data_array_x[i])*((hypothesis_function-train_data_array_y[i])**2))
                 #gradient descent theta_0 and theta_1 values

         
    
    def test_data1(self,test_data_array_x,theta_0,theta_1):
        print("In Pred : ", theta_0, theta_1)
        y_predict = [0]* len(test_data_array_x)
        length_testdata = len(test_data_array_x)
        for i in range(0,length_testdata):
            y_predict[i] = theta_0 + (theta_1 * test_data_array_x[i])
        return y_predict
       
    #calculate accuracy
    def accuracy(self,test_data_array_y, y_predict):
        print("Test_data_array_y",test_data_array_y)
        total_error = 0
        for i in range(0, len(test_data_array_y)):
            total_error += abs((y_predict[i] - test_data_array_y[i]) / test_data_array_y[i])
        total_error = (total_error / len(test_data_array_y))
        accuracy = 1 - total_error
        return accuracy * 100  
    
    
def main():
    obj = LinearRegression()
    theta_0 = 0.6
    theta_1 = 0.45

    theta_0,theta_1 = obj.gradient_descent(theta_0,theta_1,train_data_array_x,train_data_array_y)
   
    print("theta_0 value:",theta_0,"\n","theta_1 value:",theta_1)

    y_predict = obj.test_data1(test_data_array_x,theta_0,theta_1)
    print(y_predict)

    accuracy = obj.accuracy(test_data_array_y,y_predict)
    print("Accuracy", accuracy)

    

if __name__ == '__main__':
    main()
        
```

```python
import pandas as pd 
import numpy as np
```

```python
class Simple_Linear_regression:
    def __init__(self):
        self.learning_rate = 0.011
        self.epoch = 2000
        self.theta_0 = 0.6
        self.theta_1 = 0.45
    
    #load the csv file
    def function_data(self):
        self.train_data = pd.read_csv("train.csv")
        self.test_data = pd.read_csv("test.csv")
    
    #handle the missing data from test and train data
    def missing_data_handling_train(self):
        print("display datatype:\n",self.train_data.dtypes)  
        print("\n")
        print("check null value:\n", self.train_data.isnull().sum()) 
        self.train_data.replace(np.NaN, self.train_data.mean(), inplace = True) 
    
    def missing_data_handling_test(self):
        print("display datatype:\n",self.test_data.dtypes)
        print("\n")
        print("check null value:\n",self.test_data.isnull().sum()) 
        self.test_data.replace(np.NaN, self.test_data.mean(), inplace = True)
        
    #feature scalling on train data
    def feature_scalling(self):
        self.train_data = self.train_data - self.train_data.min()/self.train_data.max()-self.train_data.min()
        print("Simple feature scaling\n",self.train_data)
        return self.train_data
   
    #convert given data into array form using numpy
    def split(self):
        train_data_array_x = np.array(self.train_data.x)
        train_data_array_y = np.array(self.train_data.y)
        
        test_data_array_x = np.array(self.test_data.x)
        test_data_array_y = np.array(self.test_data.y)
        
        return train_data_array_x, train_data_array_y, test_data_array_x,test_data_array_y
    
    #find gradient descent
    def gradient_descent(self,train_data_array_x,train_data_array_y):     
        length_traindata = len(train_data_array_x)
        for value_item in range(self.epoch):
            h2 = 0
            h1 = 0
            for i in range(length_traindata):
                hypothesis_function = self.theta_0 + (self.theta_1 * train_data_array_x[i])
                h1 += (hypothesis_function - train_data_array_y[i])
                h2 += (hypothesis_function - train_data_array_y[i] * train_data_array_x[i])
#             self.theta_0 = self.theta_0 - ((self.learning_rate  * h1) / len(train_data_array_x))
#             self.theta_1 = self.theta_1 - ((self.learning_rate * h2) / len(train_data_array_x))
            self.theta_0 = self.theta_0 - ((self.learning_rate / len(train_data_array_x)) * h1)
            self.theta_1 = self.theta_1 - ((self.learning_rate  / len(train_data_array_x)) * h2)
        
        return [self.theta_0, self.theta_1]
    

    def test_data1(self, test_data_array_x, list_data1):
#        list_data1 = []
        y_predict = [0]* len(test_data_array_x)
        length_testdata = len(train_data_array_x)
        for i in range(length_testdata):
            y_predict[i] = (list_data1[0] + (list_data1[1] * test_data_array_x[i]))
            return y_predict
       
   
    #calculate accuracy
    def accuracy(self, test_data_array_y, y_predict):
        total_error = 0
        for i in range(0, len(test_data_array_y)):
            total_error += abs((y_predict[i] - test_data_array_y[i]) / test_data_array_y[i])
            total_error = (total_error / len(test_data_array_y))
        accuracy = 1 - total_error
        accuracy_result = accuracy * 100
        print("Accuracy", accuracy_result)

obj = Simple_Linear_regression()
```

```python
y_predict = obj.test_data1(test_data_array_x, list_data1)
```
